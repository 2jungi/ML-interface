```python
import os, time
import json
```

```python
def checkFile_orderParsing(checkFileName, deleteFile_TF=False):
    while (True):
        try:
            MLinputFile = open(checkFileName, "r")
            break
        except:
            print("Checking filename: {:s}...".format(checkFileName))
            time.sleep(1)
    
    fileText = MLinputFile.read()
    
    print("Parsing the file...")
    print(fileText)
    
    orderParam = {}
    for subText in fileText.split("\n"):
        subTextList = subText.split("=")
        indexStr = subTextList[0].strip()
        if len(subTextList) == 2:
            fieldStr = subTextList[1].strip()
            #print(subTextList)
            if indexStr == "seqNum": orderParam["seqNum"] = fieldStr  # all parameter are string types
            elif indexStr == "dataType": orderParam["dataType"] = fieldStr
            elif indexStr == "dataPath": orderParam["dataPath"] = fieldStr
            elif indexStr == "orderTime": orderParam["orderTime"] = fieldStr
            elif indexStr == "limitTime": orderParam["limitTime"] = fieldStr
            else: 
                #print(subTextList)
                pass
        else: 
            #print(subTextList)
            pass

    if len(orderParam) == 5: 
        pass
    else:
        print("Error in parsing the file")
        orderPara = -1
        
    os.remove(checkFileName) if deleteFile_TF == True else _
    
    return orderParam
```

```python
def makeMLresult_to_jsonFile(orderParam, resultParam):
    reportStr = orderParam
    reportStr.update(resultParam)
    #print(reportStr)
    
    filePath = "./MLoutput_" + orderParam["dataType"] + "_" + orderParam["seqNum"] + ".json"
    with open(filePath, "w") as jsonFile:  
        json.dump(reportStr, jsonFile)
        print("File {:s} is submitted".format(filePath))
        
    return 0
```

```python
def do_MLprocess_by_dataType(orderParam):
    print("MLprocess will try on the order sequence: {:s}, dataType: {:s}".format(orderParam["seqNum"], orderParam["dataType"]))
    
    dataType = orderParam["dataType"]
    if dataType == "FrontRGB_ETRI":
        print("Running ML-{:s} process running...".format(dataType))
        MLresultParam = do_MLprocess_FrontRGB_ETRI(orderParam)

    elif dataType == "Sound_ETRI":
        print("Running ML-{:s} process running...".format(dataType))
        MLresultParam = do_MLprocess_Sound_ETRI(orderParam)

    elif dataType == "Lidar_ETRI":
        print("Running ML-{:s} process running...".format(dataType))
        MLresultParam = do_MLprocess_Lidar_ETRI(orderParam)

    elif dataType == "SideRGB_Alchera":
        print("Running ML-{:s} process running...".format(dataType))
        MLresultParam = do_MLprocess_SideRGB_Alchera(orderParam)

    elif dataType == "SideIR_Alchera":
        print("Running ML-{:s} process running...".format(dataType))
        MLresultParam = do_MLprocess_SideIR_Alchera(orderParam)

    else:
        print("No ML-{:s} process...".format(dataType))
        MLresultParam = -1

    return MLresultParam
```

```python
if __name__ == "__main__":
    checkFileName = 'MLinput_ETRI.txt'

    orderParam = checkFile_orderParsing(checkFileName, deleteFile_TF=False)
    if orderParam == -1: 
        print('Error... Program terminated')
        #exit()
    else:
        pass

    MLresultParam = do_MLprocess_by_dataType(orderParam)
    if MLresultParam == -1: 
        print('Error... Program terminated')
        #exit()
    else:
        pass
    
    #check limit time, not implemented yet

    returnCode = makeMLresult_to_jsonFile(orderParam, MLresultParam)
    if returnCode == -1: 
        print('Error... Program terminated')
        #exit()
    else:
        print("###Successfully done... Program terminated", "green")
        #exit()
```
